### Notes on how this was prepared.
* The starter dataset contains all the "yes" / "no" questions from the VQA v2 training annotations.

* Each image and question is used exactly once.
* The image filenames correspond to the [COCO 2014 training images](http://images.cocodataset.org/zips/train2014.zip). Note this file is about 11GB. You will need to download this later.

````python
# Starter code to download small files
# Use this to download your starter CSV in notebook one and two.
# Also use it to download your thumbnails.zip in notebook two.
def download_if_missing(url, target, extract=True):
  if os.path.exists(target):
    return target
  return tf.keras.utils.get_file(target, origin=url, extract=extract)
````

### Get to know your data.
Write code to display a few images, questions, and answers.

````python
# GET ALL IMAGE PATHS IN THE IMAGE DIRECTORY
data_root = '/content/coco_images/'
data_root = pathlib.Path(data_root)
all_image_paths = list(data_root.glob('*/*'))
````

````python
# CREATING DATAFRAME OBJECT WITH QUESTION, ANSWER, IMAGE FILE NAME
import pandas as pd
df = pd.read_csv(csv_path, header = None, names = ['question', 'answer', 'image'])
````

````python
# CREATE LIST OF IMAGES FOR WHICH I'VE PULLED DF INFO
images_to_check = []

# WANT TO CHECK THREE, BUT ONLY IMAGES THAT ARE IN DF
while len(images_to_check) < 3:
  image_path = random.choice(all_image_paths)

  # ONLY DO REST IF IT'S IN DF
  if str(image_path)[31:] in (set(df['image'])):

    print(df[df['image'] == str(image_path)[31:]]['question'].values) #GET Q
    print(df[df['image'] == str(image_path)[31:]]['answer'].values) #GET A
    images_to_check.append(str(image_path)[31:]) #ADD TO CHECKED LIST

# SHOW IMAGE
    display.display(display.Image(image_path))
  print()
````

````python
# DOUBLE CHECK NO MISSING IMAGES

# THIS SHOULD CAPTURE ANY IMAGES THAT NOT IN ALL_IMAGE_PATHS
missing_images = [x for x in set(df['image']) if pathlib.Path('/content/coco_images/train2014/' + x) not in all_image_paths]

# AND SHOW THEM
missing_images
````

````python
# DEFINE WHERE WE ARE LOOKING
data_root = '/content/coco_images/'
data_root = pathlib.Path(data_root)
all_image_paths = list(data_root.glob('*/*'))

# GET STRINGS OF ALL IMAGES
all_image_paths = [str(path) for path in all_image_paths]

# GET STRINGS THAT ARE IN STARTER CSV
image_paths_to_keep = [x for x in all_image_paths if x[31:] in set(df['image'])]
random.shuffle(all_image_paths)

image_count = len(all_image_paths)
image_count
````

````python
# MAKE DIRECTORY OF FILES TO PUT IN ZIP
!mkdir files_to_zip 
````

````python
for img in image_paths_to_keep: #ONLY RESIZE IMAGES IN CSV

  image_file = pathlib.Path(img[9:])

  # RESIZE
  size = 299, 299
  im = Image.open(img)
  im.thumbnail(size) #RESIZE
  out_file = os.path.join('files_to_zip/', img[31:])
  im.save(out_file, "JPEG") #AND SAVE
````

````python
#MAKE ZIP FILE WITH ONLY DESIRED IMAGES
shutil.make_archive('thumbnails', 'zip', 'files_to_zip')
````
